```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/ASAP2_train_sourcetexts.csv')
df.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Download necessary NLTK data
nltk.download('punkt')
nltk.download("punkt_tab")
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')


# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    """Cleans and preprocesses text data."""
    if isinstance(text, str):
        text = re.sub(r'[^a-zA-Z\s]', '', text.lower()) # Remove non-alphabetic characters and convert to lowercase
        words = text.split()
        words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words] # Lemmatize and remove stopwords
        return ' '.join(words)
    return "" # Return empty string for non-string types

# Apply preprocessing to the 'full_text' column
df['cleaned_text'] = df['full_text'].apply(preprocess_text)

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000) # Consider the top 5000 features

# Fit and transform the cleaned text data
X = tfidf_vectorizer.fit_transform(df['cleaned_text'])

# Display the shape of the resulting matrix
print("Shape of TF-IDF matrix:", X.shape)
```

```python
df.drop(["source_text_2", "source_text_3", "source_text_4"], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Fill missing values in categorical columns with the mode
for col in ['economically_disadvantaged', 'student_disability_status', 'ell_status', 'race_ethnicity']:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Verify that there are no more missing values
df.isna().sum()
```

```python
categorical_cols = ['economically_disadvantaged', 'student_disability_status', 'ell_status', 'race_ethnicity', 'gender', 'prompt_name']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Since the scores are integers, we can treat this as a classification problem for the selected models.
# We need to define the target variable Y and the features X.
# X will be the TF-IDF features combined with the one-hot encoded categorical features.
# Y will be the 'score' column.

# First, let's align the TF-IDF features (X) with the DataFrame index.
# The TF-IDF vectorizer output is a sparse matrix, we can convert it to a DataFrame for easier concatenation.
X_tfidf_df = pd.DataFrame(X.toarray(), index=df.index, columns=tfidf_vectorizer.get_feature_names_out())

# Drop original text columns and the target variable 'score' from the main DataFrame to get other features
df_features = df.drop(columns=['essay_id', 'full_text', 'cleaned_text', 'assignment', 'source_text_1', 'score'])

# Concatenate the TF-IDF features with the other features
X_combined = pd.concat([X_tfidf_df, df_features], axis=1)

# Define the target variable
y = df['score']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42, stratify=y) # Stratify for classification

# Initialize and train the models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "Naive Bayes": MultinomialNB(), # MultinomialNB is suitable for count data or TF-IDF
    "SVM": SVC(probability=True) # probability=True allows using predict_proba
}

results = {}

for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, zero_division=0)

    results[name] = {
        "accuracy": accuracy,
        "report": report
    }

    print(f"{name} Accuracy: {accuracy:.4f}")
    print(f"{name} Classification Report:\n{report}")
    print("-" * 30)

# You can now compare the performance metrics stored in the 'results' dictionary
```

## Load and prepare data for BERT

### Subtask:
Load the dataset and prepare it for input to a BERT model. This involves tokenization and creating attention masks.

**Reasoning**:
Load the dataset using the `datasets` library, tokenize the 'full_text' column using a BERT tokenizer, and format the data for input to a BERT model.

```python
from datasets import Dataset, ClassLabel, Features, Value
from transformers import BertTokenizerFast
import pandas as pd

# Convert the pandas DataFrame to a Hugging Face Dataset without specifying features initially
dataset = Dataset.from_pandas(df)

# Map scores from 1-6 to 0-5 and rename the column to 'labels'
def map_score_to_label(examples):
    return {'labels': [score - 1 for score in examples['score']]}

dataset = dataset.map(map_score_to_label, batched=True)

# Remove the original 'score' column
dataset = dataset.remove_columns(['score'])

# Convert the new 'labels' column to ClassLabel
# Get the unique values in the 'labels' column (which are now 0-5) to define the class labels
unique_labels = sorted(list(set(dataset['labels'])))
dataset = dataset.cast_column('labels', ClassLabel(names=[str(label) for label in unique_labels]))


# Load a BERT tokenizer
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')

# Tokenize the text data
def tokenize_function(examples):
    # Ensure padding and truncation
    return tokenizer(examples['full_text'], padding='max_length', truncation=True, max_length=512)

tokenized_datasets = dataset.map(tokenize_function, batched=True)

# Prepare the data for the model
# We need to remove original text and other unnecessary columns.
columns_to_remove = ['essay_id', 'full_text', 'assignment', 'source_text_1', 'cleaned_text']
if '__index_level_0__' in tokenized_datasets.column_names:
    columns_to_remove.append('__index_level_0__')

tokenized_datasets = tokenized_datasets.remove_columns(columns_to_remove)


# Split the dataset into training and testing sets
# We will use the same split ratio as before (80% train, 20% test)
train_test_split_dataset = tokenized_datasets.train_test_split(test_size=0.2, seed=42, stratify_by_column="labels")

# Get the training and testing datasets
train_dataset = train_test_split_dataset['train']
test_dataset = train_test_split_dataset['test']

print("Prepared data for BERT:")
print("Training dataset:", train_dataset)
print("Testing dataset:", test_dataset)
```

## Load a pretrained BERT model

### Subtask:
Load a suitable pretrained BERT model for sequence classification.

**Reasoning**:
Load a pretrained BERT model for sequence classification, specifying the number of output labels based on the unique scores in the dataset.

```python
from transformers import BertForSequenceClassification

# Define the number of labels (which is the number of unique scores)
num_labels = dataset.features['labels'].num_classes
print(f"Number of labels: {num_labels}")

# Load a pretrained BERT model for sequence classification
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_labels)

print("\nBERT model loaded.")
```
